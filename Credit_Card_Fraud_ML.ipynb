```python
# Importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# Loading the dataset
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# Visualize the first 5 rows
credit_card_data.head()
```

```python
# This line give us the last 5 rows
credit_card_data.tail()
```

This data set is highly unbalanced, where:
0 --> Normal Transaction
1 --> Fraudulent Transaction

```python
# Separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for the two variables
credit_card_data.groupby('Class').mean()
```

```python
#Under-sampling
```

Build a sample dataset containing similar distribution of normal transactions and fraudulent transactions.

The number of fraudulent transactions is 169.

```python
legit_sample = legit.sample(n=169)
```

Concatenating two data frames (Legit and Fraudulent)

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
# Will show legit transactions
new_dataset.head()
```

```python
# Will show fraudulent transactions
new_dataset.tail()
```

```python
# Check the values counts
new_dataset['Class'].value_counts()
```

As we can see it is aligned (balanced).

```python
# Lets check the means
new_dataset.groupby('Class').mean()
```

(For Above) This tell us that the nature of the data has not changed (compared with the original dataset). So the difference is still here.

(For Below) Lets now split our data set into features and targets.

```python
X = new_dataset.drop(columns = 'Class', axis=1) # 1 represents columns axis
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Now we will split the data into training data and Testing data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression Model

```python
model = LogisticRegression(solver='lbfgs', max_iter=500)
```

```python
# training the Logistic Regression Model with training data
model.fit(X_train, Y_train)
```

Evaluate our model based on accuracy score

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data: ', testing_data_accuracy)
```
